```python
"""
This code will generate a csv file with country names
from disasters dataset and get latitude and longitude.
The data will be later joined with the main dataframe to
create a map in streamlit.

Timur Dzhafari
"""

import pandas as pd

df = pd.read_excel("D:/School/UNCC/projects/repos/streamlit_app/data/2000-2023 disaster around the world.xlsx", skiprows = 6)
df.head(10).to_string()
```

```python
df.columns
```

```python
!pip install geopy

from functools import partial
from geopy.geocoders import Nominatim

# first I need to generate lat and long for countries in the df
geolocator = Nominatim(user_agent="streamilt_map_disasters")

# geopy library indicates risk of 429 error. So I'll generate a dict
country_to_lat_lang = {}
list_of_countries = list(df["Country"].unique())

location = geolocator.geocode(list_of_countries[0])

n=0
for country in list_of_countries:
    n+=1
    try:
        country_to_lat_lang[country] = (
            geolocator.geocode(country).latitude,
            geolocator.geocode(country).longitude,
        )
        print(
            f"{country_to_lat_lang.get(country)} of {country}, countries left: is {len(list_of_countries) - n}"
        )
    except AttributeError as ae:
        print(f'AttributeError encountered trying to convert {country}, check out the stack \n {str(ae)}')
    except Exception as e:
        print(f'IDK what the heck that was but here is the country {country}, check out the stack \n {str(e)}')
```

```python
for country in list_of_countries:
    if country not in country_to_lat_lang.keys():
        print(country)
```

```python
df = pd.DataFrame.from_dict(country_to_lat_lang, orient='index', columns=["latitude", "longitude"])
df
```
